```python
#Import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1. Load the dataset
data = pd.read_csv('parkinsons.data.txt', delimiter=',')

# 2. It is always a good practice to eye-ball raw data to get a feel of the data in terms of number of 
# structure of the file, number of attributes, 
# types of attributes and a general idea of likely challenges in the dataset.

print(data.shape)
data.head()
```

```python
for feature in data.columns: # Loop through all columns in the dataframe
    if data[feature].dtype == 'object': # Only apply for columns with categorical strings
        data[feature] = pd.Categorical(data[feature]).codes # Replace strings with an integer
```

```python
data.info()
```

```python
X = data.drop('status', axis=1)
y = data.pop("status")
```

```python
# splitting data into training and test set for independent attributes
# 4. Split the dataset into training and test set in the ratio of 70:30 (Training:Test).
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, train_labels, test_labels = train_test_split(X, y, test_size=.30, random_state=1)
```

```python
# 5. Create the model using “entropy” method of reducing the entropy and fit it to training data.
model = DecisionTreeClassifier(criterion = 'entropy' )
```

```python
model.fit(X_train, train_labels)
```

```python
# 6. Test the model on test data and what is the accuracy achieved. Capture the predicted values and do a crosstab.
y_predict = model.predict(X_test)
print(model.score(X_train , train_labels))
print(model.score(X_test , test_labels))
print(metrics.confusion_matrix(test_labels, y_predict))
```

```python
'''
predicyted YES as YES 17 yes as no only 2
NO as NO only 2 NO as YES 38
True positive - 17, True negitive - 2
False Positive - 2, False negitive - 38
when compared below confusion metrix after regularization this is good
'''
```

```python
# 7. Use regularization parameters of max_depth, min_sample_leaf to recreate the model. 
# What is the impact on the model accuracy
reg_dt_model = DecisionTreeClassifier(criterion = 'entropy', max_depth = 5, min_samples_leaf=5)
reg_dt_model.fit(X_train, train_labels)
```

```python
y_predict = reg_dt_model.predict(X_test)
reg_dt_model.score(X_test , test_labels)
```

```python
print(metrics.confusion_matrix(test_labels, y_predict))
```

```python
# when we add max_depth = 5, min_samples_leaf=5 the accuracy got decreased
```

```python
# 8. Next implement the decision tree using Random Forest. What is the optimal number of trees that 
# gives the best result
from sklearn.ensemble import RandomForestClassifier
rfcl = RandomForestClassifier(n_estimators = 70)
rfcl = rfcl.fit(X_train, train_labels)
```

```python
y_predict = rfcl.predict(X_test)
print(rfcl.score(X_test , test_labels))
print(metrics.confusion_matrix(test_labels, y_predict))
```
