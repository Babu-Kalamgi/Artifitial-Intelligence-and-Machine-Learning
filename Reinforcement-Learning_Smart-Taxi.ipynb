By Using Q-Learning Algorithem

```
Q = np.zeros([n_states, n_actions])
Q
```

```
episodes = 2000
G = 0 # Total reward
NoTimeSteps = 0
alpha = 0.618
```

```
for episode in range(1, episodes+1):
  done = False
  G, reward = 0, 0
  state = env.reset()
  firstSate = state
  while reward != 20:
    action = np.argmax(Q[state])
    state2, reward, done, info = env.step(action)
    Q[state, action] += alpha * (reward + np.max(Q[state2]) - Q[state, action] )
    G += reward
    NoTimeSteps = NoTimeSteps + 1
    state = state2
```

```
print("steps : {} rewards : {}".format(NoTimeSteps, reward))
```
