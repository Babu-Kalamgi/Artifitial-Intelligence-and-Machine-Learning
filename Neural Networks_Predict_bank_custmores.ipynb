```
df = pd.read_csv(project_path + "bank.csv")
```

```
df.shape
```

```
df.head()
```

```
df = df.drop(['CustomerId','RowNumber','Surname'],axis=1)
```

```
X=df.drop("Exited",axis=1)
```

```
X.shape
```

```
Y=df[['Exited']]
```

```
Y.shape
```

```
from sklearn.preprocessing import LabelEncoder
labelencoder_x = LabelEncoder()
X.loc[:, 'Geography'] = labelencoder_x.fit_transform(X.loc[:, 'Geography'])
X.loc[:, 'Gender'] = labelencoder_x.fit_transform(X.loc[:, 'Gender'])
X.head()
```

```
from keras.utils import to_categorical
encoded = pd.DataFrame(to_categorical(X.loc[:, 'Geography']))
X = pd.concat([encoded, X], axis = 1)
X.head()
```

```
X=X.drop(['Geography',0],axis=1)
```

```
X.head()
```

```
X.shape
```

```
from scipy import stats
from scipy.stats import zscore
X.apply(zscore)
```

```
from sklearn.model_selection import train_test_split
v_size = 0.30
seed = 0
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=v_size, random_state=seed)
```

```
from keras.models import Sequential
from keras.layers import Dense 

classifier = Sequential()
classifier.add(Dense(output_dim = 6, init = 'uniform', activation = 'relu', input_dim = 11))
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))

classifier.compile(optimizer = 'adam', loss = "binary_crossentropy", metrics = ['accuracy'])
```

```
classifier.fit(X_train, Y_train, batch_size = 11, nb_epoch = 20)
```

```
Y_pred = classifier.predict(X_test)
Y_pred
```

```
Y_pred2 = Y_pred > 0.5
Y_pred2
```

```
Yp=  classifier.evaluate(X_test,Y_test)
print(Yp)
print("Accuracy of predicted value: ", Yp[1])
```

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred2)
print (cm, '\n\n', Y_pred2[:5, :])
```
